```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/HarshRaj2717/Python-Projects/master/EDA%20on%20Banglore%20house%20pricing%20dataset/Bangalore.csv")
df
```

```python
df.info
```

```python
plt.scatter(df["Price"], df["Area"])
plt.xticks(rotation=90)
plt.ylabel("Area")
plt.xlabel("Price")
```

```python
area = df.iloc[0:, 1:2]
price = df.iloc[0:, 0:1]
```

```python
area
```

```python
price
```

```python
model = LinearRegression()
model.fit(area, price)
```

```python
price_pred = model.predict(area)
price_pred
```

```python
plt.scatter(area, price)
plt.plot(area, price_pred, color="red")
plt.title("Best Fit Line")
plt.xlabel("Area")
plt.ylabel("Price")
```
